```python
import pandas as pd
from math import pi
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Vardy = {'Goals':19,'xGoals':14.2,'Penalties':4,'Shots/90':2.31,'Assists':4,'xAssists':3.4}
Messi = {'Goals':21,'xGoals':14,'Penalties':4,'Shots/90':4.7,'Assists':14,'xAssists':11.4}

data = pd.DataFrame([Messi,Vardy], index = ["Messi","Vardy"])
data
```

```python
Attributes =list(data)
AttNo = len(Attributes)

values = data.iloc[1].tolist()
values += values [:1]
values
```

```python
angles = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles += angles [:1]
values2 = data.iloc[0].tolist()
values2 += values2 [:1]
values2
```

```python
angles2 = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles2 += angles2 [:1]
```

```python
ax = plt.subplot(111, polar=True)

plt.xticks(angles[:-1],Attributes)

ax.plot(angles,values, 'royalblue', alpha = 1)
ax.fill(angles, values, 'royalblue', alpha=0.4)

ax.plot(angles2,values2, 'gold')
ax.fill(angles2, values2, 'gold', alpha=0.1)

#Rather than use a title, individual text points are added
plt.figtext(0.1,0.9,"Messi",color="gold")
plt.figtext(0.1,0.85,"v")
plt.figtext(0.1,0.8,"Vardy",color="royalblue")
plt.show()
```
