```python
import pandas as pd
from math import pi
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
vvd = {'Tackles/90':0.97,'Blocks/90':0.86,'Clearances/90':7.45,'Interceptions/90':0.90,'Aerial Duels Won/90':4.67,'Fouls/90':0.34}
ramos = {'Tackles/90':1.82,'Blocks/90':1.26,'Clearances/90':3.32,'Interceptions/90':1.46,'Aerial Duels Won/90':2.02,'Fouls/90':1.46}

data = pd.DataFrame([ramos,vvd], index = ["ramos","vvd"])
data
```

```python
Attributes =list(data)
AttNo = len(Attributes)

values = data.iloc[1].tolist()
values += values [:1]
values
```

```python
angles = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles += angles [:1]
```

```python
values2 = data.iloc[0].tolist()
values2 += values2 [:1]
values2
```

```python
angles2 = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles2 += angles2 [:1]
```

```python
ax = plt.subplot(111, polar=True)

plt.xticks(angles[:-1],Attributes)

ax.plot(angles,values, 'red')
ax.fill(angles, values, 'red', alpha=0.1)

ax.plot(angles2,values2, 'indigo')
ax.fill(angles2, values2, 'indigo', alpha=0.1)

#Rather than use a title, individual text points are added
plt.figtext(0.1,0.9,"Sergio Ramos",color="indigo")
plt.figtext(0.1,0.85,"v")
plt.figtext(0.1,0.8,"Van Dijk",color="red")
plt.show()
```
