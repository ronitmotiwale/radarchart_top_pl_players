```python
import pandas as pd
from math import pi
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Alisson = {'GoalsAgainst/90':0.57,'xGoalsAgainst/90':0.8,'Clean Sheet %':0.5,'Save %':0.8,'Saves/90':2.33}
Oblak = {'GoalsAgainst/90':0.79,'xGoalsAgainst/90':0.68,'Clean Sheet %':0.39,'Save %':0.73,'Saves/90':2.06}

data = pd.DataFrame([Oblak,Alisson], index = ["Oblak","Alisson"])
data
```

```python
Attributes =list(data)
AttNo = len(Attributes)

values = data.iloc[1].tolist()
values += values [:1]
values
```

```python
angles = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles += angles [:1]
values2 = data.iloc[0].tolist()
values2 += values2 [:1]
values2
```

```python
angles2 = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles2 += angles2 [:1]
```

```python
ax = plt.subplot(111, polar=True)

plt.xticks(angles[:-1],Attributes)

ax.plot(angles,values, 'red')
ax.fill(angles, values, 'red', alpha=0.1)

ax.plot(angles2,values2, 'darkgreen')
ax.fill(angles2, values2, 'darkgreen', alpha=0.1)

#Rather than use a title, individual text points are added
plt.figtext(0.1,0.9,"Oblak",color="darkgreen")
plt.figtext(0.1,0.85,"v")
plt.figtext(0.1,0.8,"Alisson",color="red")
plt.show()
```
