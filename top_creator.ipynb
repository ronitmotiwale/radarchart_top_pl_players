```python
import pandas as pd
from math import pi
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
DeBruyne = {'Assists':17,'xAssists':13.4,'Key Passes/90':3.85,'ThroughBalls/90':0.92,'Passes Into Final 3rd/90':5.65,'Crosses/90':10.2}
Muller = {'Assists':20,'xAssists':11.4,'Key Passes/90':3.26,'ThroughBalls/90':0.3,'Passes Into Final 3rd/90':3.67,'Crosses/90':4.39}

data = pd.DataFrame([Muller,DeBruyne], index = ["Muller","DeBruyne"])
data
```

```python
Attributes =list(data)
AttNo = len(Attributes)

values = data.iloc[1].tolist()
values += values [:1]
values
```

```python
angles = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles += angles [:1]
```

```python
values2 = data.iloc[0].tolist()
values2 += values2 [:1]
```

```python
values2
```

```python
angles2 = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles2 += angles2 [:1]
```

```python
ax = plt.subplot(111, polar=True)

plt.xticks(angles[:-1],Attributes)

ax.plot(angles,values, 'deepskyblue')
ax.fill(angles, values, 'skyblue', alpha=0.1)

ax.plot(angles2,values2, 'red')
ax.fill(angles2, values2, 'red', alpha=0.1)

#Rather than use a title, individual text points are added
plt.figtext(0.1,0.9,"Thomas Muller",color="red")
plt.figtext(0.1,0.85,"v")
plt.figtext(0.1,0.8,"Kevin De Bruyne",color="deepskyblue")
plt.show()
```
